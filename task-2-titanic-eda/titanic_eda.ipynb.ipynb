```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# General info
df.info()

# Summary stats
df.describe()

# Check missing values
df.isnull().sum()
```

```python
# Histogram of numerical features
df.hist(bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()

# Boxplots for Age and Fare
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(y='Age', data=df)
plt.title('Boxplot of Age')

plt.subplot(1, 2, 2)
sns.boxplot(y='Fare', data=df)
plt.title('Boxplot of Fare')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Pairplot 
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']])
plt.show()
```

## Observations:
- Fare has strong outliers (seen in boxplot).
- Age is right-skewed and varies a lot.
- Pclass is negatively correlated with Fare and Survived.
- Survivors mostly paid higher fares and were in higher classes.
- Some correlation between SibSp and Parch (family size).
